```python
# 5 id diferentes

ID = data.iloc[ :5, 0]
print(ID)

ID.to_csv("puntoA.csv")
```

```python
# Host despúes del 2020

Host2020 = data[(data["host_since"] >= '01/01/20')]
Host2020.to_csv("puntoB.csv")
Host2020
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
data['host_since']= data['host_since'].astype('datetime64[s]')
```

```python
# Usuarios con más de 100 reseñas

Propiedades = data[(data["number_of_reviews"] > 100)]
Propiedades.to_csv("puntoC.csv")
Propiedades
```

```python
# Filas multiplos de 200
multiplos200 = data.iloc[::200]
multiplos200.to_csv("puntoD.csv")
multiplos200
```

```python
# cuartos tipo "Entire home" y "private room"
TipoCuarto = data[["calculated_host_listings_count_entire_homes", "calculated_host_listings_count_private_rooms"]]
TipoCuarto.to_csv("puntoE.csv")
TipoCuarto
```

```python
# Registros con 4 camas o menos
Camas = data[data["beds"] <= 4]
Camas.to_csv("puntoF.csv")
Camas
```

```python
# Registros con la isignia Superhost
Superhost = data[data["host_is_superhost"] == "t"]
Superhost.to_csv("puntoG.csv")
Superhost
```
